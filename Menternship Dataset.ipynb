# 2) Multivariate Analysis

```python
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=df, color='red')
plt.show()
```

```python
selected_cols = ['BMI', 'MentHlth', 'PhysHlth', 'Age', 'Income']
sns.pairplot(df[selected_cols])
plt.show()
```

```python
pd.crosstab(df['HighBP'], df['Stroke']).plot(kind='bar', stacked=True)
plt.show()
```

```python
sns.countplot(x='Smoker', hue='HeartDiseaseorAttack', data=df)
plt.show()
```

```python
corr = df[['HeartDiseaseorAttack', 'Smoker', 'Stroke', 'Age']]
sns.heatmap(corr.corr(), annot=True , cmap='coolwarm')
plt.show()
```
